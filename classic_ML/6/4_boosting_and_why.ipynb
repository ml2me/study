```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
e = np.random.normal(0,1/5,300)
```

```python
x = np.linspace(1, 6, 300)
```

```python
df = pd.DataFrame()
df['y'] = np.cos(x) + e
df['x'] = x
```

```python
df.plot(x='x', y='y', kind='scatter', figsize=(15,10))
```

```python
df['pred0'] = df['y'].mean()
```

```python
ax = df.plot(x='x', y='y', kind='scatter', figsize=(15,10))
df.plot(x='x', y='pred0', kind='scatter', ax=ax, color='r')
```

```python
df['res0'] = df['y'] - df['pred0']
```

```python
df
```

```python
df['x'].values.reshape(-1,1)
```

```python
from sklearn.tree import DecisionTreeRegressor
reg1 = DecisionTreeRegressor(max_depth=1)
reg1.fit(df['x'].values.reshape(-1, 1), df['res0']) 
tree_pred1 = reg1.predict(df['x'].values.reshape(-1, 1))
```

```python
df['tree_pred1'] = tree_pred1
```

```python
df
```

```python
ax = df.plot(x='x', y='y', kind='scatter', figsize=(15,10))
df.plot(x='x', y='res0', kind='scatter', ax=ax, color='r')
df.plot(x='x', y='tree_pred1', kind='scatter', ax=ax, color='g')
```

```python
learning_rate = 0.3
df['pred1'] = df['pred0'] + learning_rate * df['tree_pred1']
```

```python
df
```

```python
ax = df.plot(x='x', y='y', kind='scatter', figsize=(15,10))
df.plot(x='x', y='pred0', kind='scatter', ax=ax, color='r')
df.plot(x='x', y='pred1', kind='scatter', ax=ax, color='g')
```

```python
df['res1'] = df['y'] - df['pred1']
```

```python
df
```

```python
reg2 = DecisionTreeRegressor(max_depth=1)
reg2.fit(df['x'].values.reshape(-1, 1), df['res1']) 
tree_pred2 = reg1.predict(df['x'].values.reshape(-1, 1))
```

```python
df['tree_pred2'] = tree_pred2
df['pred2'] = df['pred1'] + learning_rate * df['tree_pred1']
df
```

```python
ax = df.plot(x='x', y='y', kind='scatter', figsize=(15,10))
df.plot(x='x', y='pred0', kind='scatter', ax=ax, color='b')
df.plot(x='x', y='pred1', kind='scatter', ax=ax, color='r')
df.plot(x='x', y='pred2', kind='scatter', ax=ax, color='g')
```

```python
trees = 10
trees_list = []
df['pred'] = df['y'].mean()

for i in range(trees):
    df['res'] = df['y'] - df['pred']
    reg = DecisionTreeRegressor(max_depth = 1)
    reg.fit(df['x'].values.reshape(-1, 1), df['res']) 
    tree_pred = reg.predict(df['x'].values.reshape(-1, 1))
    df['tree_pred'] = tree_pred
    df['pred'] = df['pred'] + learning_rate * df['tree_pred']
    
    ax = df.plot(x='x', y='y', kind='scatter', figsize=(15,10))
    df.plot(x='x', y='pred', kind='scatter', ax=ax, color='r', title = str(i+1))
    
```

```python
from sklearn.metrics import mean_squared_error
```

```python
df['true'] = np.cos(x)
```

```python
trees = 131
trees_list = []
df['pred'] = df['y'].mean()

for i in range(trees):
    df['res'] = df['y'] - df['pred']
    reg = DecisionTreeRegressor(max_depth = 1)
    reg.fit(df['x'].values.reshape(-1, 1), df['res']) 
    tree_pred = reg.predict(df['x'].values.reshape(-1, 1))
    df['tree_pred'] = tree_pred
    df['pred'] = df['pred'] + learning_rate * df['tree_pred']
    
    if i % 10 == 0:
        
        MSE = round(mean_squared_error(df['y'], df['pred']), 4)
        MSE_true = round(mean_squared_error(df['true'], df['pred']), 4)
        
        print(i, 'MSE =', MSE)
        
        ax = df.plot(x='x', y='y', kind='scatter', figsize=(15,10))
        df.plot(x='x', y='true', kind='scatter', ax=ax, color='g')
        df.plot(x='x', y='pred', kind='scatter', ax=ax, color='r', title = str(i+1)+', MSE '+str(MSE)+', MSE_true '+str(MSE_true))
        
```

```python
for depth in range(1,10):

    big_tree = DecisionTreeRegressor(max_depth = depth)
    big_tree.fit(df['x'].values.reshape(-1, 1), df['y']) 
    big_tree_pred = big_tree.predict(df['x'].values.reshape(-1, 1))
    df['big_tree_pred'] = big_tree_pred

    MSE = round(mean_squared_error(df['y'], df['big_tree_pred']), 4)
    MSE_true = round(mean_squared_error(df['true'], df['big_tree_pred']), 4)
    
    print(depth, 'MSE =', MSE)
    
#    if (depth == 5) | (depth == 6) | (depth == 9):
    if (depth > -1):
                
        ax = df.plot(x='x', y='y', kind='scatter', figsize=(15,10))
        df.plot(x='x', y='true', kind='scatter', ax=ax, color='g')
        df.plot(x='x', y='big_tree_pred', kind='scatter', ax=ax, color='r', title = str(depth)+', MSE '+str(MSE)+', MSE_true '+str(MSE_true))
```
