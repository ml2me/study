```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
20*649   # оценка размера (потом можно будет только например 2026 год парсить)
```

```python
# 'https://verenitsa.ru/blagodarim-za-pozhertvovanie?wpv_view_count=7923&wpv_paged=1'

urls = [str(i) for i in range(649)]

data = []
for url in urls:
    response = requests.get('https://verenitsa.ru/blagodarim-za-pozhertvovanie?wpv_view_count=7923&wpv_paged='+url)
    # print(url)  # отладочный вывод
    soup = BeautifulSoup(response.text, 'html.parser')

    # Ищем div с результатами
    results_div = soup.find('div', {'id': 'wpv-view-layout-7923'})

    if not results_div:
        print(f"Не найден div с результатами: https://verenitsa.ru/blagodarim-za-pozhertvovanie?wpv_view_count=7923&wpv_paged={url}")
        continue

    # Ищем все строки таблицы внутри tbody
    tbody = results_div.find('tbody', {'class': 'wpv-loop js-wpv-loop'})

    if not tbody:
        print(f"Не найден tbody: https://verenitsa.ru/blagodarim-za-pozhertvovanie?wpv_view_count=7923&wpv_paged={url}")
        continue

    rows = tbody.find_all('tr')

    for row in rows[1:]: # пропускаем заголовочную строку
        cols = row.find_all('td')
        if len(cols) >= 5:  # проверяем, что есть все нужные колонки
            date = cols[0].text.strip()
            name = cols[1].text.strip()
            what_4 = cols[2].text.strip()
            what_way = cols[3].text.strip()
            # Преобразуем сумму в число, убирая пробелы
            amount_str = cols[4].text.strip().replace(' ', '')
            try:
                amount = float(amount_str)
            except ValueError:
                amount = 0.0

            data.append([date, name, what_4, what_way, amount])
            # print(f"Добавлена запись: {date}, {name}, {amount}")  # отладочный вывод
print('done')
```

```python
print(f"Всего записей: {len(data)}")
df = pd.DataFrame(data, columns=['date', 'name', 'what_4', 'what_way', 'amount'])
df.to_csv('output.csv', index=False, encoding='utf-8-sig')
print("Данные сохранены в output.csv")
```
