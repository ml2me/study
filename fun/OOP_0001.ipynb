```python
class Point():
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __repr__(self):
        return f'Point(x={self.x!r}, y={self.y!r})'
```

```python
test_point = Point(1, 2)
print(test_point)
print(eval(repr(test_point)) == test_point) # False (Разные id)

# Чтобы eval(repr(obj)) == obj возвращало True, нужно определить метод __eq__ 
```

```python
class Item():
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    def __repr__(self):
        return f"Item({self.name!r}, {self.price!r})"
    
    def __lt__(self, other):
        return self.price < other.price
```

```python
test_item = Item('juehtw', 333)
print(test_item)

test_item_2 = eval(repr(test_item))
print(test_item_2)

test_item_3 = Item('ju', 334)
print(test_item_3)

print(sorted([test_item, test_item_2, test_item_3], reverse=True))
```
