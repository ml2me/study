```python
from datetime import date, datetime

def count_age(birthday):
    today = datetime.now().date()
    # Основная разница лет
    age = today.year - birthday.year
    # Уменьшаем на 1, если день рождения еще не наступил в этом году
    if (today.month, today.day) < (birthday.month, birthday.day):
        age -= 1
    print(f"age: {age}, запас {75 - age}" )

count_age(date(1962, 3, 19))
count_age(date(1959, 4, 20))
```
